```python
df['MultipleLines'] = df.MultipleLines.apply(lambda x: x if len(x) < 5 else 'No') 
df['OnlineSecurity'] = df.OnlineSecurity.apply(lambda x: x if len(x) < 5 else 'No') 
df['OnlineBackup'] = df.OnlineBackup.apply(lambda x: x if len(x) < 5 else 'No') 
df['DeviceProtection'] = df.DeviceProtection.apply(lambda x: x if len(x) < 5 else 'No') 
df['TechSupport'] = df.TechSupport.apply(lambda x: x if len(x) < 5 else 'No') 
df['StreamingMovies'] = df.StreamingMovies.apply(lambda x: x if len(x) < 5 else 'No') 
df['MultipleLines'] = df.MultipleLines.apply(lambda x: x if len(x) < 5 else 'No') 
df['StreamingTV'] = df.StreamingTV.apply(lambda x: x if len(x) < 5 else 'No') 
```

```python
df.Churn.value_counts()
```

```python
for col in df.columns:
    if max([len(str(i)) for i in df[f"{col}"].value_counts().keys()]) > 5:
        print(col)
```

```python
df['MonthlyCharges'] = pd.to_numeric(df['MonthlyCharges'], errors='coerce')
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Iterate and update the DataFrame
for i in range(len(df)):
    if pd.isna(df.at[i, 'TotalCharges']):
        df.at[i, 'TotalCharges'] = df.at[i, 'MonthlyCharges'] * 12
```

```python
def all_integers(seriesObject):
    for val in seriesObject:
        if not isinstance(val, int) :
            print(val, type(val))
            return False
    return True
```
